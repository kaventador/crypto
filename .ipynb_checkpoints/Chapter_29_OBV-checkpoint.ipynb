```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = yf.download(tickers='XRP-USD',start='2023-01-10')
df
```

```python
def OBV(DF):
    df = DF.copy()
    df['daily ret'] = df['Close'].pct_change()
    df['direction'] = np.where(df['daily ret']>0 , 1 ,-1)
    df['direction'][0] = 0
    df['adj volume'] = df['Volume']*df['direction']
    df['obv'] = df['adj volume'].cumsum()
    return df['obv']
```

```python
data = OBV(df)
data
```

```python
plt.figure(figsize=(16,8))
ax1 = plt.subplot2grid((11,1),(0,0),rowspan= 5 , colspan= 1)
ax2 = plt.subplot2grid((11,1),(6,0),rowspan= 5 , colspan= 1)

ax1.plot(df['Close'],linewidth = 2 , label ="PRICE" , color='#ff9800')
ax1.set_title('Closing Price')
ax2.plot(data,linewidth = 3 , label ="OBV" , color='#26a69a',alpha=0.3)
ax2.set_title('OBV INDICATOR')



ax1.legend()
ax2.legend()
plt.show()
```
