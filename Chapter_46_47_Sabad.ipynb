```python
import pandas as pd
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
```

```python

assets = ['BTC-USD' , 'ETH-USD' , 'ADA-USD' , 'TRX-USD' , 'BNB-USD' , 'XRP-USD' , 'MATIC-USD' , 'LINK-USD' , 'LTC-USD']
data = pd.DataFrame()
for a in assets:
    data[a+' Close'] = yf.download(tickers = a , start='2020-01-01' , interval='1mo')['Close']
    data[a+' Volume'] = yf.download(tickers = a , start='2020-01-01' , interval='1mo')['Volume']
```

```python
data
```

```python
return_df = pd.DataFrame()
for a in assets:
    print('Calculating Monthly Return: ' , a)
    data[a+ ' month_ret']=data[a+' Close'].pct_change()
    return_df[a] = data[a+ ' month_ret']
```

```python
return_df
```

```python
def CAGR(DF):
    df = DF.copy()
    df["cum_return"] = (1 + df["mon_ret"]).cumprod()
    n = len(df)/12
    CAGR = (df["cum_return"].tolist()[-1])**(1/n) - 1
    return CAGR

def volatility(DF):
    df = DF.copy()
    vol = df["mon_ret"].std() * np.sqrt(12)
    return vol

def sharpe(DF,rf):
    df = DF.copy()
    sr = (CAGR(df) - rf)/volatility(df)
    return sr
    
def max_dd(DF):
    df = DF.copy()
    df["cum_return"] = (1 + df["mon_ret"]).cumprod()
    df["cum_roll_max"] = df["cum_return"].cummax()
    df["drawdown"] = df["cum_roll_max"] - df["cum_return"]
    df["drawdown_pct"] = df["drawdown"]/df["cum_roll_max"]
    max_dd = df["drawdown_pct"].max()
    return max_dd
```

```python

def pflio(DF,m,x):
    df = DF.copy()
    portfolio = []
    monthly_ret = [0]
    for i in range(1,len(df)):
        if len(portfolio) > 0:
            monthly_ret.append(df[portfolio].iloc[i,:].mean())
            bad_stocks = df[portfolio].iloc[i,:].sort_values(ascending=True)[:x].index.values.tolist()
            portfolio = [t for t in portfolio if t not in bad_stocks]
        fill = m - len(portfolio)
        new_picks = df.iloc[i,:].sort_values(ascending=False)[:fill].index.values.tolist()
        portfolio = portfolio + new_picks
        print(portfolio)
    monthly_ret_df = pd.DataFrame(np.array(monthly_ret),columns=["mon_ret"])
    return monthly_ret_df
```

```python
portfo_ret = pflio(return_df,6,3)
```
