```python
import pandas as pd
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
```

```python
assets = ['BTC-USD' , 'ETH-USD', 'ADA-USD' , 'TRX-USD' , 'BNB-USD' , 'XRP-USD' , 'MATIC-USD' , 'LINK-USD' , 'LTC-USD']
```

```python

data = pd.DataFrame()
for a in assets:
    data[a+' Close'] = yf.download(tickers = a , start= '2023-01-01')['Close']
    data[a+' High'] = yf.download(tickers = a , start= '2023-01-01')['High']
    data[a+' Low'] = yf.download(tickers = a , start= '2023-01-01')['Low']
    data[a+' Volume'] = yf.download(tickers = a , start= '2023-01-01')['Volume']
```

```python
data
```

```python
def ATR(DF,i,n):
    df = DF.copy()
    df[i+'H-L']=abs(df[i+' High']- df[i+' Low'])
    df[i+'H-PC']=abs(df[i+' High']-df[i+' Close'].shift(1))
    df[i+'L-PC']=abs(df[i+' Low']-df[i+' Close'].shift(1))
    df[i+'TR']=df[[i+'H-L', i+'H-PC', i+'L-PC']].max(axis=1,skipna=False)
    df[i+' ATR'] = df[i+'TR'].rolling(n).mean()
    df2 = df.drop([i+'H-L',i+'H-PC',i+'L-PC'],axis=1)
    return df2[i+' ATR']
```

```python
tickers_signal = {}
tickers_ret = {}
```

```python
for a in assets:
    print("calculating ATR and rolling max price for ",a)
    data[a+" ATR"] = ATR(data,a,20)
    data[a+" roll_max_cp"] = data[a+" High"].rolling(20).max()
    data[a+" roll_min_cp"] = data[a+" Low"].rolling(20).min()
    data[a+" roll_max_vol"] = data[a+" Volume"].rolling(20).max()
    data.dropna(inplace=True)
    tickers_signal[a] = ""
    tickers_ret[a] = []
```

```python
data
```

```python
for a in assets:
    print("calculating returns for ",a)
    for i in range(len(data)):
        if tickers_signal[a] == "":
            tickers_ret[a].append(0)
            if data[a+" High"][i]>=data[a+" roll_max_cp"][i] and                data[a+" Volume"][i]>1.5*data[a+" roll_max_vol"][i-1]:
                tickers_signal[a] = "Buy"
            elif data[a+" Low"][i]<=data[a+" roll_min_cp"][i] and                data[a+" Volume"][i]>1.5*data[a+" roll_max_vol"][i-1]:
                tickers_signal[a] = "Sell"
        
        elif tickers_signal[a] == "Buy":
            if data[a+' Close'][i]<data[a+' Close'][i-1] - data[a+" ATR"][i-1]:
                tickers_signal[a] = ""
                tickers_ret[a].append(((data[a+' Close'][i-1] - data[a+" ATR"][i-1])/data[a+' Close'][i-1])-1)
            elif data[a+" Low"][i]<=data[a+" roll_min_cp"][i] and                data[a+" Volume"][i]>1.5*data[a+" roll_max_vol"][i-1]:
                tickers_signal[a] = "Sell"
                tickers_ret[a].append(((data[a+' Close'][i-1] - data[a+" ATR"][i-1])/data[a+' Close'][i-1])-1)
            else:
                tickers_ret[a].append((data[a+' Close'][i]/data[a+' Close'][i-1])-1)
                
        elif tickers_signal[a] == "Sell":
            if data[a+' Close'][i]>data[a+' Close'][i-1] + data[a+" ATR"][i-1]:
                tickers_signal[a] = ""
                tickers_ret[a].append((data[a+' Close'][i-1]/(data[a+' Close'][i-1] + data[a+" ATR"][i-1]))-1)
            elif data[a+" High"][i]>=data[a+" roll_max_cp"][i] and                data[a+" Volume"][i]>1.5*data[a+" roll_max_vol"][i-1]:
                tickers_signal[a] = "Buy"
                tickers_ret[a].append((data[a+' Close'][i-1]/(data[a+' Close'][i-1] + data[a+" ATR"][i-1]))-1)
            else:
                tickers_ret[a].append((data[a+' Close'][i-1]/data[a+' Close'][i])-1)
                
    data[a+" ret"] = np.array(tickers_ret[a])
```

```python
for a in assets:
    print(a + ": " +tickers_signal[a])
```

```python

strategy_df = pd.DataFrame()
for a in assets:
    strategy_df[a] = data[a+" ret"]
strategy_df["ret"] = strategy_df.mean(axis=1)

strategy_df
```
